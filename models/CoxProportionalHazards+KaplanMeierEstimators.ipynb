```python
# Kaplan-Meier Survival Curve
from lifelines import KaplanMeierFitter
from lifelines.utils import median_survival_times
import pandas as pd

# Read datafiles
train_file_path = "1yVTNHjwM4hUbv0alTfW9hEn1itOVbHLg"
test_file_path = "1-1NGzpvsdLcBPFZU_aNOZYn-1BIMoWO2"
train_url = f"https://drive.google.com/uc?export=download&id={train_file_path}"
test_url = f"https://drive.google.com/uc?export=download&id={test_file_path}"
train_df = pd.read_csv(train_url)
test_df = pd.read_csv(test_url)

# Instantiate KaplanMeierFitter
kmf = KaplanMeierFitter()

#Save  time variable(efs_time) in T and event/status variable (efs) in E
T = train_df["efs_time"]
E = train_df["efs"]

# Drop columns with high collinearity
train_df = train_df.drop('gvhd_proph_FK+- others(not MMF,MTX)', axis=1)

# Fit the Kaplan-Meier model
kmf.fit(durations=T, event_observed=E)

# Print Median Survival Time and Confidence Intervals
median_ = kmf.median_survival_time_
median_confidence_interval_ = median_survival_times(kmf.confidence_interval_)

print(f"Median Survival Time: {median_:.2f} months")
print(f"95% Confidence Interval for Median Survival Time: {median_confidence_interval_}")

'''
Interpretation:
- You can be 95% confident that the true median survival time for the population lies between 9.664 and 10.31 months.
- If you were to repeat the study with different samples, 95% of those samples would show a median survival time within this range.
'''

# Cox Proportional Hazards Model
from lifelines import CoxPHFitter

# Instantiate CoxPHFitter with a penalizer for Lasso regularization
cph = CoxPHFitter(penalizer=0.1)

# Fit the Cox model on the training data
cph.fit(train_df, duration_col='efs_time', event_col='efs')

# Print the CoxPH model summary
cph.print_summary()

'''
Interpretation:
- The Cox Proportional Hazards model can help us understand the relationship between covariates (e.g., age, treatment type) and survival.
- The coefficients in the summary will tell us the direction and strength of the relationship between each covariate and the risk of an event occurring (death, relapse, etc.).
'''
```
